```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt

sb.set(rc={'figure.figsize':(15,8)})
```

```python
df_escolas = pd.read_csv('/content/alunos_rede estadual_2023_1SEMESTRE.csv', sep=';')
df_escolas.head(20)
```

```python
df_escolas.info()
```

```python
# vamos fazer um histograma para visualizar o nº de matrículas em anos iniciais
sb.displot(df_escolas['ANOS INICIAIS'])
plt.show()
```

```python
# vamos excluir escolas que não oferece anos iniciais para os estudantes, criando um filtro

filtro_escolas = df_escolas[df_escolas['ANOS INICIAIS'] > 0]
sb.displot(filtro_escolas['ANOS INICIAIS'], bins=10)
plt.show()
```

Vemos a distribuição no eixo x da quantidade de matriculados nos anos iniciais e no eixo y a quantidade de escolas. A maioria das escolas possui até 750 alunos matriculados nos anos iniciais.

```python
# vamos olhar para as escolas que oferecem os anos finais

filtro_escolas = df_escolas[df_escolas['ANOS FINAIS'] > 0]
sb.displot(filtro_escolas['ANOS FINAIS'], bins=10)
plt.show()
```

A concentração de escolas que possuem anos finais é maior do que anos iniciais. A concentração maior é até 400 alunos matriculados por escola, chegando até 1200 escolas.

```python
fundamental_iniciais = df_escolas['ANOS INICIAIS'].sum()
fundamental_finais = df_escolas['ANOS FINAIS'].sum()
ensino_medio = df_escolas['ENSINO MÉDIO'].sum()
(fundamental_iniciais + fundamental_finais + ensino_medio)
```

Temos 3.101.719 alunos matriculados nos três níveis de ensino da rede estadual.

```python
# vamos criar uma lista para fazer um gráfico de setores/pizza
alunos = [fundamental_iniciais, fundamental_finais, ensino_medio]
periodo = ['fundamental anos iniciais', 'fundamental anos finais', 'ensino médio']

plt.pie(alunos, labels=periodo, autopct='%0.0f%%')
plt.show()
```

```python
filtro_escolas = df_escolas[df_escolas['ENSINO MÉDIO'] > 0]
sb.boxplot(filtro_escolas['ENSINO MÉDIO'])
plt.show()
```

```python
# vamos olhar a média de alunos matriculados nos anos finais do ensino fundamental

df_escolas['ANOS FINAIS'].describe()
```

```python
# vamos olhar para escolas que possuem alunos matriculados tanto nos anos iniciais quanto finais do ensino fundamental

filtro_escolas = df_escolas[(df_escolas['ANOS INICIAIS'] > 0) & (df_escolas['ANOS FINAIS'] > 0)]
sb.scatterplot(x=filtro_escolas['ANOS INICIAIS'], y=filtro_escolas['ANOS FINAIS'])
plt.show
```

Há uma boa correlação até a quantidade de 600 alunos matriculados. Isso pode ser explicado, pois a maioria das escolas possuem os dois níveis do ensino fundamental.
